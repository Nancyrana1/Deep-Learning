COMPARISON CODE OF CPU AND GPU

```python
import time

size=10000

a = torch.randn(size,size)
b = torch.randn(size,size)

start_time = time.time()
matrix_multi_cpu = torch.matmul(a,b)
print("CPU time - > ",time.time() - start_time)

a1 = a.to('cuda')
b1 = b.to('cuda')

start_time = time.time()
matrix_multi_gpu = torch.matmul(a1,b1)
print("GPU time - > ",time.time() - start_time)
```
